```python
df['gender'].unique()
gender_map = {
    'M':'Male','m':'Male','males':'Male','mal':'Male',
    'F':'Female','f':'Female','females':'Female','femal':'Female'
}
df['gender'] = df['gender'].str.lower().map(
    lambda x: gender_map.get(x, x.title()) if pd.notna(x) else x
)
```

```python
caste_map = {
    'maratha':'Maratha','marathas':'Maratha','marata':'Maratha',
    'muslim':'Muslim','mulsim':'Muslim',
    'other obc':'Other OBC','otherobc':'Other OBC','othr obc':'Other OBC','other_obc':'Other OBC',
    'sc':'SC','s.c.':'SC',
    'st':'ST','s.t.':'ST',
    'nt_vj':'NT_VJ','nt|vj':'NT_VJ',
    'kunbi':'Kunbi','other general':'Other General','others':'Others'
}
valid_castes = ['SC','ST','NT_VJ','Muslim','Kunbi','Other OBC','Maratha','Other General','Others']
df['caste'] = (df['caste']
    .str.strip()
    .str.lower()
    .map(lambda x: caste_map.get(x, x) if pd.notna(x) else x)
)
df['caste'] = df['caste'].where(df['caste'].isin(valid_castes), other=np.nan)
```

```python
df.loc[~df['age'].between(18, 100), 'age'] = np.nan
band_range = {'18-25':(18,25),'26-35':(26,35),'36-50':(36,50),'51-60':(51,60),'61+':(61,100)}
def band_ok(row):
    if pd.isna(row['age']) or pd.isna(row['age_band']):
        return True
    lo, hi = band_range.get(row['age_band'], (0, 200))
    return lo <= row['age'] <= hi
mismatch = ~df.apply(band_ok, axis=1)
df.loc[mismatch, 'age'] = np.nan
```

```python
MY_PARTIES  = ['BJP','SHS (Shinde)','NCP (AP)','JSS/RSVA/RYSP']
MVA_PARTIES = ['CONG','SHS (UBT)','NCP (SP)','SP/PWPI']
def alliance_ok(row):
    a, p = row['voted_alliance'], row['voted_party']
    if a == 'Maha Yuti' and p not in MY_PARTIES:  return False
    if a == 'MVA'and p not in MVA_PARTIES: return False
    if a == 'VBA' and p != 'VBA':           return False
    return True
bad_match = ~df.apply(alliance_ok, axis=1)
print(f'Alliance/party mismatches set to NaN: {bad_match.sum()}')
df.loc[bad_match, 'voted_party'] = np.nan
```

```python
mum_rural = (df['region'] == 'Mumbai') & (df['geography'] == 'Rural')
print(f'Mumbai-Rural fixed: {mum_rural.sum()}')
df.loc[mum_rural, 'geography'] = 'Urban'
print(f'\nFinal shape after cleaning: {df.shape}')
print('Remaining nulls:\n', df.isnull().sum())
```

```python
for col in ['gender','caste','occupation','cm_preference']:
    df[col].fillna(df[col].mode()[0], inplace=True)
df['age'] = df['age'].fillna(
    df.groupby('age_band')['age'].transform('mean').astype(np.int64)
)
df['voting_reason'].fillna('Unknown', inplace=True)
```
